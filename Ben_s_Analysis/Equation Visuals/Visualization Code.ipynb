```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("./../Output/LowPPMMatrix.csv")

# Setting the aesthetics for the plots
sns.set(style="whitegrid")

# Plotting histograms for the independent variables: 'Resistance', 'RelativeHumidity', and 'Temperature'
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(18, 5))

sns.histplot(data['Resistance'], bins=30, ax=axes[0], color='skyblue')
axes[0].set_title('Resistance Distribution')

sns.histplot(data['RelativeHumidity'], bins=30, ax=axes[1], color='lightgreen')
axes[1].set_title('Relative Humidity Distribution')

sns.histplot(data['Temperature'], bins=30, ax=axes[2], color='salmon')
axes[2].set_title('Temperature Distribution')

plt.tight_layout()
plt.show()

# Also, a scatter plot of the dependent variable 'Target PPM' against each independent variable
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(18, 5))

sns.scatterplot(x=data['Resistance'], y=data['Target PPM'], ax=axes[0], color='skyblue')
axes[0].set_title('Target PPM vs Resistance')

sns.scatterplot(x=data['RelativeHumidity'], y=data['Target PPM'], ax=axes[1], color='lightgreen')
axes[1].set_title('Target PPM vs Relative Humidity')

sns.scatterplot(x=data['Temperature'], y=data['Target PPM'], ax=axes[2], color='salmon')
axes[2].set_title('Target PPM vs Temperature')

plt.tight_layout()
plt.show()
```
